```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np

from bs4 import BeautifulSoup
import requests
```

```python
url='http://www.vpgt.lt/vpgt/m/m_events_summary/wfiles/e_2018-08-01.htm'

result = requests.get(url)
c = result.content
soup =BeautifulSoup(c,'lxml')
table = soup.find('table')
```

```python
rows =table.find_all('tr')
data = []
for tr in rows:
    cols = tr.findAll('td')
    
    for td in cols:
        text1 = td.find('p')
        data.append(text1.text)
```

```python
data=data[data.index('Ekstremaliąsias situacijas'):data.index('2.\r\n  Gaisrų aprašymas')]
```

```python
df=pd.DataFrame(np.array(data).reshape(22,14), columns =["Duomenys apie:", "2017 nuo metų pr.", "2018 nuo metų pr.", "Iš viso per parą", "Vilniaus", "Kauno", "Klaipėdos", "Šiaulių", "Panevėžio", "Alytaus", "Utenos", "Marijampolės", "Tauragės", "Telšių"])
```

```python
df=df.drop(df.columns[1:4], axis=1).T.drop([2,6,13], axis=1)
```

```python
df=df.rename(columns=df.iloc[0]).drop(['Duomenys apie:']).reset_index()
```
