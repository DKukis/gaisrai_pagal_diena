```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np

from bs4 import BeautifulSoup
import requests

from datetime import timedelta, date

import itertools
```

```python
datadate=[]

def daterange(date1, date2):
    for n in range(int ((date2 - date1).days)+1):
        yield date1 + timedelta(n)

start_dt = date(2015, 12, 24)
end_dt = date(2015, 12, 25)


for dt in daterange(start_dt, end_dt):
    datadate.append(dt.strftime("%Y-%m-%d"))
    
datadate
```

```python
data=[]
dfs=[]
headers=[]

for datos in datadate:
    url='http://www.vpgt.lt/vpgt/m/m_events_summary/wfiles/e_'+datos+'.htm'
    result = requests.get(url)
    c = result.content
    soup =BeautifulSoup(c,'lxml')
    table = soup.find('table')
    rows =table.find_all('tr')
    
    data0=[]
    
    for tr in rows:
        cols = tr.find_all('td')
        for td in cols:
            text = td.find_all(text=True)
            for t in text:
                data0.append(t)
                    
    data0=data0[data0.index('Ekstremaliąsias situacijas'):data0.index('2.\r\n  Gaisrų aprašymas')]
    data0 = [a for a in data0 if a != '\n']
    data0[140:143] = [''.join(data0[140:143])]
    
    df=pd.DataFrame(np.array(data0).reshape(22,14), columns =["Duomenys apie:", "2017 nuo metų pr.", "2018 nuo metų pr.", "Iš viso per parą", "Vilniaus", "Kauno", "Klaipėdos", "Šiaulių", "Panevėžio", "Alytaus", "Utenos", "Marijampolės", "Tauragės", "Telšių"])
    df=df.drop(df.columns[1:4], axis=1)
    df['Data']=datos
    #df=df.set_index('Duomenys apie:')
    dfs.append(df)
    #headers.append(datos)                    
    
    #data.append(data0)
    
result = pd.concat(dfs, ignore_index=True) 
```
