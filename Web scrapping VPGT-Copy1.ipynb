```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np

from bs4 import BeautifulSoup
import requests

from datetime import timedelta, date
```

```python
datadate=[]

def daterange(date1, date2):
    for n in range(int ((date2 - date1).days)+1):
        yield date1 + timedelta(n)

start_dt = date(2015, 12, 20)
end_dt = date(2015, 12, 21)


for dt in daterange(start_dt, end_dt):
    datadate.append(dt.strftime("%Y-%m-%d"))
    
datadate
```

```python
data=[]

for datos in datadate:
    url='http://www.vpgt.lt/vpgt/m/m_events_summary/wfiles/e_'+datos+'.htm'
    result = requests.get(url)
    c = result.content
    soup =BeautifulSoup(c,'lxml')
    table = soup.find('table')
    rows =table.find_all('tr')
    data0=[]
    for tr in rows:
        cols = tr.find_all('td')
        for td in cols:
            text1 = td.find('p')
            data0.append(text1.text)
            
    data.append(data0)
```
